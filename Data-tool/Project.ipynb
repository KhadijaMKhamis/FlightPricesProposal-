```python
flight.columns = [
    'leg_id',
    'search_date',
    'flight_date',
    'starting_airport',
    'destination_airport',
    'fare_basis_code',
    'travel_duration',
    'elapsed_days',
    'is_basic_economy',
    'is_refundable',
    'is_non_stop',
    'base_fare',
    'total_fare',
    'seats_remaining',
    'total_travel_distance',
    'departure_epoch',
    'departure_time',
    'arrival_epoch',
    'arrival_time',
    'arrival_airport_code',
    'departure_airport_code',
    'airline_name',
    'airline_code',
    'equipment_description',
    'duration_seconds',
    'segment_distance',
    'cabin_code'
]

flight.to_csv("flight_renamed.csv", index=False)
```

```python
flight.columns
```

```python
flight["flight_date"] = pd.to_datetime(flight["flight_date"])
flights_per_day = flight["flight_date"].value_counts().sort_index()

print(flights_per_day)
```

```python
dates = flights_per_day.index
counts = flights_per_day.values

plt.plot(dates, counts, marker='o')
plt.title("Number of Flights per Day in May")
plt.xlabel("Date")
plt.ylabel("Number of Flights")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
departures_count = flight["departure_airport_code"].value_counts().head(10)

print(departures_count)
```

```python
departures_count.plot(kind='bar')

plt.title("Top 10 Departure Airports in May")
plt.xlabel("Departure Airport")
plt.ylabel("Number of Flights")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
arrival_count = flight["arrival_airport_code"].value_counts().head(10)

print(arrival_count)
```

```python
arrival_count.plot(kind='bar')

plt.title("Top 10 Arrival Airports in May")
plt.xlabel("Arrival Airport")
plt.ylabel("Number of Flights")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
ord_departures = flight[flight["departure_airport_code"] == "ORD"]
ord_flights_per_day = ord_departures["flight_date"].value_counts().sort_index()
for date, count in ord_flights_per_day.items():
    
    print(f"{date.date()} --- {count} flights")
```

```python
ord_flights_per_day.plot(kind='line', marker='o')

plt.title("Number of Flights Departing from ORD per Day (May)")
plt.xlabel("Date")
plt.ylabel("Number of Flights")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
ord_departures = ord_departures.copy()
ord_departures["fare_bhd"] = ord_departures["total_fare"] * 0.376

cheapest_daily = ord_departures.loc[ord_departures.groupby("flight_date")["total_fare"].idxmin()]
summary_table = cheapest_daily[["flight_date", "airline_name", "total_fare", "fare_bhd"]].copy()
summary_table[["total_fare", "fare_bhd"]] = summary_table[["total_fare", "fare_bhd"]].round(2)
summary_table = summary_table.sort_values("flight_date").reset_index(drop=True)

print(summary_table)
```

```python
plt.figure(figsize=(12,6))
plt.plot(summary_table["flight_date"], summary_table["fare_bhd"], marker='o')
plt.title("Cheapest Daily Fare (May) from ORD – in BHD")
plt.xlabel("Date")
plt.ylabel("Fare (BHD)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
ord_departures["day_of_week"] = pd.to_datetime(ord_departures["flight_date"]).dt.day_name()
cheapest = ord_departures.loc[ord_departures.groupby("day_of_week")["total_fare"].idxmin()]
cheapest = cheapest[["day_of_week", "airline_name", "total_fare", "fare_bhd"]]
cheapest = cheapest.set_index("day_of_week").loc[[
    "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"
]]

print(cheapest)
```

```python
cheapest["fare_bhd"].plot(kind='bar')

plt.title("Cheapest Fare per Day from ORD (BHD)")
plt.xlabel("Day")
plt.ylabel("Fare (BHD)")
plt.show()
```

```python
from IPython.display import display

ord_departures["day_of_week"] = ord_departures["flight_date"].dt.day_name()
cheapest = ord_departures.loc[ord_departures.groupby("day_of_week")["total_fare"].idxmin()]
cheapest["days_between"] = (pd.to_datetime(cheapest["flight_date"]) - pd.to_datetime(cheapest["search_date"])).dt.days
weekday_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
cheapest["day_of_week"] = pd.Categorical(cheapest["day_of_week"], categories=weekday_order, ordered=True)
cheapest = cheapest.sort_values("day_of_week").reset_index(drop=True)
display(cheapest[[
    "day_of_week",
    "search_date",
    "flight_date",
    "days_between",
    "airline_name",
    "departure_airport_code",
    "arrival_airport_code",
    "is_non_stop",
    "cabin_code",
    "total_fare",
    "fare_bhd"
]])
```
